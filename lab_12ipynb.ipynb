```python
generator = pipeline(
    "text-generation",
    model="gpt2"
)
```

```python
prompt1 = "Machine learning is"

output1 = generator(
    prompt1,
    max_length=40,
    num_return_sequences=1
)

print("Prompt 1 Output:\n")
print(output1[0]['generated_text'])
```

```python
prompt2 = "Generative AI can be used for"

output2 = generator(
    prompt2,
    max_length=40
)

print("\nPrompt 2 Output:\n")
print(output2[0]['generated_text'])
```

```python
prompt3 = "The future of artificial intelligence is"

output3 = generator(
    prompt3,
    max_length=40
)

print("\nPrompt 3 Output:\n")
print(output3[0]['generated_text'])
```
