```python
# Import libraries
import pandas as pd
import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer

# Download required data (IMPORTANT)
nltk.download('punkt')
nltk.download('punkt_tab')   # Fixes your current error
nltk.download('stopwords')

# Sample dataset
data = {
    'text': [
        "Machine learning is amazing",
        "Natural language processing is useful",
        "Text preprocessing improves results",
        "Deep learning models are powerful"
    ]
}

df = pd.DataFrame(data)

# Lowercase
df['text'] = df['text'].str.lower()

# Tokenization
df['tokens'] = df['text'].apply(word_tokenize)

# Remove stopwords
stop_words = set(stopwords.words('english'))

df['clean'] = df['tokens'].apply(
    lambda words: [w for w in words if w not in stop_words]
)

# Stemming
stemmer = PorterStemmer()

df['stemmed'] = df['clean'].apply(
    lambda words: [stemmer.stem(w) for w in words]
)

# Vectorization
vectorizer = CountVectorizer()

X = vectorizer.fit_transform(df['text'])

# Output
print("\nProcessed Data:")
print(df)

print("\nVectorized Output:")
print(X.toarray())
```
